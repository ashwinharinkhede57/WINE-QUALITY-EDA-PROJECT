```python
import pandas as pd
import seaborn as sns


df = pd.read_csv("/content/winequality-red.csv", encoding= 'unicode_escape')
```

```python
df["alcohol"].value_counts()
```

```python
df.describe ()
```

```python
df['alcohol'].min(),df['alcohol'].max()
```

```python
plt.figure(figsize=(10,5))
sns.distplot(df['alcohol'],color="k")
plt.title('Distplot for winequality ',fontsize=10)
plt.xlabel('alcohol',fontsize=10)
plt.ylabel('Density',fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(x='alcohol',data=df,color="r")
plt.title('Histogram for alcohol',fontsize=10)
plt.xlabel('alcohol',fontsize=10)
plt.ylabel('Density',fontsize=10)
plt.show()
```

```python
plt.figure()
sns.boxplot(x='quality', y='alcohol', data=df, palette='cool')
plt.title('Alcohol content by Wine Quality')
plt.show()
```

```python
plt.figure()
sns.countplot(x='quality', data=df, palette='magma')
plt.title('Count of Wine Quality Ratings')
plt.show()
```

```python
plt.figure()
sns.violinplot(x='quality', y='sulphates', data=df, palette='cubehelix')
plt.title('Sulphates by Wine Quality')
plt.show()
```

```python
plt.figure()
sns.kdeplot(df['pH'], shade=True, color='crimson')
plt.title('pH Density Plot')
plt.show()
```

```python
plt.figure()
sns.lineplot(x='quality', y='citric acid', data=df, color='teal')
plt.title('Fixed Acidity vs Citric Acid')
plt.show()
```

```python
sns.boxplot(x=df['pH'], color='darkgreen')
plt.title('Boxplot of pH')
plt.show()
```

```python
sns.kdeplot(df['alcohol'], color='purple', shade=True)
plt.title('KDE Plot of Alcohol')
plt.show()
```

```python
sns.kdeplot(df['chlorides'], color='purple', shade=True)
plt.title('KDE Plot of Alcohol')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['alcohol'], df['pH'], s=df['quality']*20, c=df['quality'], cmap='plasma', alpha=0.6)
plt.colorbar(label='Quality')
plt.xlabel('Alcohol')
plt.ylabel('pH')
plt.title('Bubble Chart: Alcohol vs pH vs Quality')
plt.show()
```
